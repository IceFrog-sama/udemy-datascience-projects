```python
import pandas as pd
```

```python
raw_csv = pd.read_csv('Absenteeism_data.csv')
```

```python
df = raw_csv.copy()
```

```python
df
```

```python
df = df.drop(['ID'], axis=1)
```

```python
df
```

```python
reason_dummies = pd.get_dummies(df['Reason for Absence'],drop_first=True)
reason_dummies
```

```python
reason_dummies['check'] = reason_dummies.sum(axis=1)
reason_dummies
```

```python
reason_dummies = reason_dummies.drop(['check'], axis=1)
reason_dummies
```

```python
reason_dummies.iloc[:, 20:]
```

```python
group1 = reason_dummies.iloc[:, 0:14].max(axis=1)
group2 = reason_dummies.iloc[:, 14:17].max(axis=1)
group3 = reason_dummies.iloc[:, 17:20].max(axis=1)
group4 = reason_dummies.iloc[:, 20:].max(axis=1)
group4
```

```python
df_grouped_reasons = pd.concat([df, group1, group2, group3, group4], axis=1)
df_grouped_reasons
```

```python
df_grouped_reasons.columns.values
```

```python
df_grouped_reasons.columns = ['Reason for Absence', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
df_grouped_reasons
```

```python
df_reordered = df_grouped_reasons[['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Reason for Absence', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours']]
df_reordered
```

```python
df_reordered = df_reordered.drop(['Reason for Absence'], axis=1)
df_reordered
```

```python
df_reordered['Date'] = pd.to_datetime(df_reordered['Date'], format='%d/%m/%Y')
df_reordered
```

```python
def getMonth(dateVal):
    return dateVal.month
```

```python
df_reordered['Month Value'] = df_reordered['Date'].apply(getMonth)
df_reordered
```

```python
def getDay(dateVal):
    return dateVal.weekday()
```

```python
df_reordered['Day of the Week'] = df_reordered['Date'].apply(getDay)
df_reordered
```

```python
df_with_dates = df_reordered.drop(['Date'], axis=1)
df_with_dates
```

```python
df_with_dates.columns.values
```

```python
df_with_dates = df_with_dates[['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value',
       'Day of the Week', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']]
df_with_dates
```

```python
df_with_dates['Education'].unique()
```

```python
df_with_dates['Education'] = df_with_dates['Education'].map({1:0, 2:1, 3:1, 4:1})
df_with_dates
```

```python
df_preprocessed = df_with_dates.copy()
```

```python
df_preprocessed.to_csv('Absenteeism_preprocessed.csv', index=False)
```
